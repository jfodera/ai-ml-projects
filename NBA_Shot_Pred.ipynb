```python


# Getting shot CSV's from drive folder specified above and converting to data frame
playerCsvFiles = glob.glob(os.path.join(folder_path, '*.csv'))       # finds all .csv files
playerCsvFiles = sorted(playerCsvFiles)

allPlayerDataFrames = []
for f in playerCsvFiles:
    df = pd.read_csv(f)
    allPlayerDataFrames.append(df)

allShotDataFrames = []
for f in shotsCsvFiles:
    df = pd.read_csv(f)
    allShotDataFrames.append(df)







```

### Back to Boosting
- To find weak rule, we apply base learning ML algs with a different distribution
- each time base learning is applied, it generates a new weak prediction rule
- *Prediction rule* - what a node is split on
- After many iterations, the boosting algorithm combines the weak rules into a single prediction rule

#### Different types
- Adaptive Boosting (AdaBoost)
  - Steps:
    - Build a model and make predictions
    - assign higher weights to miss-classified points.
    - Build next model.
    - repeat steps 3 and 4
    - make a final model using the weighted average of the induvidual models
- Gradient Tree Boosting
  - combines prediction of multiple weak learners (typically decision trees) sequentually. It aims to improve overall performance by optimizing model weights based on errors of previous iterations.
  - mostly used for regression type problems in comparison to classification problems
- XGBoost (remember attributes from previous paper as they are not included here)
  - **Weighted quantile Sketch** - XGBoost has a distributed weighted quantile sketch algorithms to effectively handle weighted data
    - **Quantile Sketch** - clever summary of data instead of checking every single row
  - **Cache awareness** - non-continuous memory access is required to get the gradient statistics by row index.
    - XGBoosed makes optimal use of hardware by allocating internal buffers in each thread, where gradient statistics can be stores
  - **Out of Core Computing** - Optimizes available disk space and maximizes its usage when handling huge datasets that do not fit into memory
  - interpretability goes slightly down

- These methods are used in a lot of different domains showed in various research papers
- SHAP is basically a Mathmatical framework that is sometimes standard (especially in decision trees) and helps increase interpretability of the results.

## Nueral Networks
### Motivation
- our brain is an extremly large interconnected network of neurons that model the world around us
- the neuron collects inputs from other neurons using dendrites
- neuron then sums up all the inputs, if the resulting value is greater than a certain threshold, it fires
  - axon then sends this signal to other neurons


### Definition
- aka artificial neural network (ANNS)
- heart of deep learning algorithms
- ANNs are composed of node layers:
  - input layer
  - one or more hidden layers
  - output layer
- each node connects to another and has an associated weight and threshold
  - these weights and thresholds decide what information can be passed to the next layer

### Classification and Regression Trees

- CART algorithim is as follows:
- begin with root node, containing the complete dataset
- find best attribute in the data set using one of the Attribute Selection Measures (ASM)
- Divide the dataset into subsets that contain possible values for the best attribute(s)
- Generate the decision tree node that has the best attribute included in the boolean logic
- Make new decision trees using subset ofhte dataset created in step 3
- continue this process until a stage is reached where you cannot further classify the nodes, call the final nodes leaf nodes





#### How to split in tree based algorithm?
- Biggest decision is selecting the best attribute
  - Use Attribute Selection Measures (ASM)
- Selection is done by measuring entropy
  - *Entropy* - measure of uncertainty or lack of predictability associated with random variable drawn from given distribution
- Main Techniques:
  - Information Gain
  - Gini Index

#### Information Gain
- Entropy - measure of disorganizaiton in a system
  - sample completey homogeneous, entropy is zero
  - sample is equally divided (50%-50%) entropy is one
- $Entropy(S)= -p log_2p -q log_2q$
  - *p* - probability of success
    - The number of instances of the first class divided by the total number of instances in that node
  - *q* - probability of failure
    - The number of instances of the second class divided by the total number of instances.
- In this alg, we choose the split with the lowest Entropy as compared to the parent node.
- A node having the highest information gain is split first.
- Information gain formula: $$IG = Entropy(Parent) - \sum \left( \frac{N_{child}}{N_{parent}} \times Entropy(Child) \right)$$
- Good example of how to use this formula:
  1. **Identify the Parent:** Look at your total data before the split ($N_{parent}$).
  2. **Split by Feature:** Pick a feature (e.g., "Color").
  3. **Find the Children:** If "Color" has three values (Red, Blue, Green), you now have **three child nodes**.
  4. **Calculate Weighted Entropy:**
  * **Weight:** What % of the total data is Red? (e.g., $0.5$)
  * **Entropy:** How "mixed" is the Red group? (e.g., $0.8$)
  * **Multiply:** $0.5 \times 0.8 = 0.4$


  5. **Sum them up:** Do this for Blue and Green as well, then add those results together.
  6. **Subtract from Parent:** Take your original Parent Entropy and subtract that sum.


#### Gini Index
- low Gini Index, means better atttribute
- $$Gini_{Attribute} = \sum_{i=1}^{n} \left( \frac{N_{child\_i}}{N_{parent}} \times Gini(Child\_i) \right)$$
  - as seem in handwritten notes
- We calculate the Gini Index for each attribute, then pick the one with the lowest one
  - you are technically calculate the gini index of the attribute and condition `age < 50`
    - the value is typically a mean or median
- At each step, we want the node to have a gini index of zero
  - meaning each split outputs a single class 100% of the time

### Overfitting
- Decision trees (especially regression) are prone to overfitting.
- **Overfitting** - when ML model fits training data too tight and therefore fails to generalize on the test data
- Typically occurs due to the following reasonsL
  - training data size is too small, not containing enough samples to accurately represent all possible value
  - training data has a lot of noise (irrelevant information)
  - model trains for too long on a single sample set of data
  - model complexity is high, so it learns the noise in the training data
    - made worse if training data is extra noisy.

## Ensemble Learning
- a single decision tree could suffer from high bias, or high variance
- to mitigate this, we use **Ensemble Learning**
- These models make predictions by combining more than one model. Resulting in:
  - more flexibility (less bias)
  - less sensitivity (less variance)
- Popular Ensemble Methods
  - *Bagging* - training a number of individual models in a parallel manner
    - each model is trained by a random subset of data
    - aka 'Bootstrap Aggregation'
  - *Stacking* - fitting many different model types on the same data. Then using another model to learn how to best combine predictions from the other models
  - *Boosting* - Training a number of individual models in a sequential way. With each one leearning from the mistakes of the previous

### Bagging
- seeks a diverse group of weak learners by varying training data.
- In a classification setting, the alg fits base classifiers each on random subset of the original data.
  - it then aggregates their individual predictions (by voting or by averaging) to form a final prediction
- The **key** is the manner in which each sample of the dataset is selected to train the ensemble memers
  - each model gets its own unique sample
- **Bootstrap Sample** - rows are drawn from dataset at random with replacement.
  - same observation may be selected more than once or never.
  - common stats technique where if statistical estimate of a quantity is required, it can be done by creating bootstrap samples, estimate the quantity and compute the mean of the estimates
- Similarly we can prepare multiple training datasets
  - these would be independent of eachother
  - also the more ideal, and accurate scenario, but typically we do not have enough data to go with thins method, so we do bootstrapping.
  - The final result is either the average of all results or the majority vote (classification problem).
- Some popular ensemble methods that use this approach:
  - Bagged decision trees
  - Random Forest



#### Bagging sklearn Implementation:
- *n_estimators* - when set to 50
  - creates 50 bootstrap samples and 50 independant trees

```python
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import BaggingClassifier
# define toy dataset
X, y = make_classification(n_samples=1000, n_features=20, n_informative=15, n_redundant=5, random_state=5)
print(X.shape, y.shape)
# define the model
model = BaggingClassifier(n_estimators=50)
# evaluate the model
## Defines CV Strategy- this fits the model to the data, but only 90% of it, it is a practice run
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
## actually Evaluates defined model using cross validation
n_scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))

# fit the model on the whole dataset
model.fit(X, y)

# make a single prediction
row = [[-4.7705504,-1.88685058,-0.96057964,2.53850317,-6.5843005,3.45711663,-7.46225013,2.01338213,-0.45086384,-1.89314931,-2.90675203,-0.21214568,-0.9623956,3.93862591,0.06276375,0.33964269,4.0835676,1.31423977,-2.17983117,3.1047287]]
yhat = model.predict(row)
print('Predicted Class: %d' % yhat[0])
```

### Stacking
- Varies the model types fit on the training data and uses a model to process outputs and combine predictions
- Linear model is most commonly used to aggregate predictions, although you can use any model.
- It is encouraged to have the complexity reside in the lower-level ensemble member models and the
- Typically (but not always) yields performance better than any one of the singularly trained models
- Base Models
  - usually 2 or more
  - referred to as "level-0 models"
- Meta Model
  - combines the predictions of the base
  - reffered to as "level-1 model"



- training data is used to train the level-0 models.
- The predictions from the base models are made then these predictions, along with the expected outputs are the input and output pairs used to train the meta model

#### Setting up Stacking Sklearn Example
- can set meta model using the `final_estimator` argument
- meta model dataset prepared via cross validation
  - 5 fold cv by default
  - can be changed via `cv` argument
- Sequence when running `stacking.fit(X,y)`:


##### 1. The CV "Preparation" Phase (Before Final Weights)

The models have **not** established their final weights yet. Instead, the `StackingClassifier` starts the CV process:

* **The Split:** It splits your data into, say, 5 folds.
* **The Training:** It takes the **first 4 folds** and trains "temporary" versions of your LR and SVM. *These models establish weights based only on those 4 folds.*
* **The Prediction:** These temporary models predict the **5th fold**.
* **The Cycle:** This repeats until every row has an "out-of-sample" prediction.

##### 2. The Level-1 "Meta" Phase (Setting Ensemble Weights)

Now that the CV has finished, the **Meta-Model** (the "Manager") looks at those predictions and the true labels ($y$).

* It establishes **its own weights** (e.g., "I trust the SVM's opinion 70% of the time and the LR's opinion 30% of the time").

##### 3. The Final "Level-0" Phase (Establishing Final Weights)

**This is the part that answers your question.** Only *after* the Meta-Model has learned how to combine the opinions does the `StackingClassifier` go back to the very beginning.

* It takes the **entire 100% of the dataset**.
* It trains the LR and SVM one last time.
* **Now**, and only now, do the Level-0 models establish their **final, "production-ready" weights.**

- So as a summary: The Level-0 is trained twice, once on CV as a simulation run so that it doesn't see all of the data, and next with all of the data after the meta weights are set and **finalized**.

- Sometimes, better performance occurs if the rows with the original data that was fed into the level-0 model are naturally joined with the output of level-0, so that during training, the meta model has even more information
  - one can do this by simply setting the `passthrough=True`. It is not enabled by default

### Boosting
- **weak learner** - model that is only slightly better then random guessing
- ensemble meta-algoritm for primarily reducing bias, and also variance in supervised learning
- Boosting represents a family of machine learning algorithms that sequentially works on itself, converting weak learners to strong ones.
- adds multiple weak learners in an adaptive way
  - it is a generic algorithm rather than a specific model
  - requires you to specify a weak model (regression, shallow decision trees) and then improves it
- popular boosting algorithms
  - AdaBoost
  - Gradient Boost
- XGBoost - optimized version of gradient boosting algorithm
  - XGBoost has ability to penalize complex models through both L1 and L2 regularizaiton, helping to preventing overfitting.
    - *L1 Regularization (Lasso)* - adds penalty equal to abs val of the magnitude of the weights
    - *L2 Regularization (Ridge)* - adds penalty equal to the square magnitude of the weights   
  - XGBoost incorporates a sparsity-aware split finding algorithm that handles different types of sparsity patterns in the data to help combat missing values
  - *Block structure for parallel learning* - XGBoost can make use of multiple cores on the CPU
    - possible thanks to the block structure in its system design.
    - data is sorted and stored in in-memory units called blocks. Unlike other algorithms, this enables the data layout to be reused by subsequent iterations instead of recomputing
      - teq is also useful for 'split finding' and 'column sub-sampling'
